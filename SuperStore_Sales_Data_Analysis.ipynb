Filling null values

check duplicates


```python
if(data.duplicated().sum() > 0):
  print('there are duplicates')
else:
  print('no duplicates')
```

Types of Customers

```python
types_of_customers = data['Segment'].unique()

num_of_customers = data['Segment'].value_counts().reset_index()
num_of_customers.columns = ['Type Of Customers','Total']
print(types_of_customers)
print(num_of_customers)
```

```python
plt.pie(num_of_customers['Total'],labels=num_of_customers['Type Of Customers'],autopct='%1.2f%%')
plt.show()
```

```python
sales_per_segment = data.groupby('Segment')['Sales'].sum().reset_index()
sales_per_segment.columns = ['Type Of Customers','Total Sales']
print(sales_per_segment)
```

```python
plt.bar(sales_per_segment['Type Of Customers'],sales_per_segment['Total Sales'])
plt.show()
```

```python
plt.pie(sales_per_segment['Total Sales'],labels=sales_per_segment['Type Of Customers'],autopct='%1.2f%%')
plt.show()
```

```python
customer_order_frequency = data.groupby(['Customer ID', 'Customer Name','Segment'])['Order ID'].count().reset_index()
customer_order_frequency.columns = ['Customer ID','Customer Name','Type of customers','Order Frequency']

repeat_customers = customer_order_frequency[customer_order_frequency['Order Frequency'] >= 10]
repeat_customers_sorted =  repeat_customers.sort_values(by='Order Frequency',ascending=False)
repeat_customers_sorted.head(6)
```

```python
customer_sales = data.groupby(['Customer ID', 'Customer Name','Segment'])['Sales'].sum().reset_index()
customer_sales.columns = ['Customer ID','Customer Name','Type of customers','Total Sales']

top_spenders =  customer_sales.sort_values(by='Total Sales',ascending=False)
top_spenders.head(6).reset_index(drop=True)
```

```python
shipping_model = data['Ship Mode'].value_counts().reset_index()
shipping_model.columns = ['Shipping Model','Total']
shipping_model
```

```python
plt.pie(shipping_model['Total'],labels=shipping_model['Shipping Model'],autopct='%1.2f%%')
plt.show()
```

```python
state = data['State'].value_counts().reset_index()
state.columns = ['State','Total']
state.head()
```

```python
city  = data['City'].value_counts().reset_index()
city.columns = ['City','Total']
city.head()
```

```python
state_sales = data.groupby('State')['Sales'].sum().reset_index()
state_sales.columns = ['State','Total Sales']

top_states_sales = state_sales.sort_values(by='Total Sales',ascending=False)
top_states_sales.head().reset_index(drop=True)
```

```python
city_sales = data.groupby('City')['Sales'].sum().reset_index()
city_sales.columns = ['City','Total Sales']

top_cities_sales = city_sales.sort_values(by='Total Sales',ascending=False)
top_cities_sales.head().reset_index(drop=True)
```

```python
import plotly.express as px
data_summary = data.groupby(['Category','Sub-Category'])['Sales'].sum().reset_index()
fig = px.treemap(data_summary,path=['Category','Sub-Category'],values='Sales')
fig.show()
fig = px.sunburst(data_summary,path=['Category','Sub-Category'],values='Sales')
fig.show()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'],dayfirst=True)
yearly_sales = data.groupby(data['Order Date'].dt.year)['Sales'].sum()
yearly_sales = yearly_sales.reset_index()
yearly_sales.columns = ['Year','Total Sales']
yearly_sales
```

```python
# plt.bar(yearly_sales['Year'],yearly_sales['Total Sales'])
plt.plot(yearly_sales['Year'],yearly_sales['Total Sales'],marker='o',linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'],dayfirst=True)
year_sales = data[data['Order Date'].dt.year==2018]
quarterly_sales = year_sales.resample('Q',on='Order Date')['Sales'].sum()
quarterly_sales = quarterly_sales.reset_index()
quarterly_sales.columns = ['Quarter','Total Sales']
quarterly_sales
```

```python
plt.plot(quarterly_sales['Quarter'],quarterly_sales['Total Sales'],marker='o',linestyle='--')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.xticks(rotation=75)
plt.show()
```

```python
Monthly_sales = year_sales.resample('ME',on='Order Date')['Sales'].sum()
Monthly_sales = Monthly_sales.reset_index()
Monthly_sales.columns = ['Month','Total Sales']
Monthly_sales
```

```python
plt.plot(Monthly_sales['Month'],Monthly_sales['Total Sales'],marker='o',linestyle='--')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
